## data inputs

```python
import pandas as pd
sample_submission = pd.read_csv("../input/digit-recognizer/sample_submission.csv")
test = pd.read_csv("../input/digit-recognizer/test.csv")
train = pd.read_csv("../input/digit-recognizer/train.csv")
```

## dataset Analysis

```python
#length of train
print("Number of train samples", len(train))
#length of test
print("Number of test samples", len(test))
```

```python
train.head()
```

```python
def give(train):
    y = train['label']
    X = train.drop(['label'], axis=1)
    return X,y
```

```python
#reference : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html
y = train['label']
X = train.drop(['label'], axis=1)
X.head()
```

```python
# number of classes
set(y), len(set(y))
```

```python
#train -dev split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
# length of new_train
print("number of sample in train", len(X_train))
print("number of sample in test", len(X_test))
```

## Baseline

```python
#training
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(100, 20,10), random_state=1)
clf.fit(X_train, y_train)
print("model trained--")
#predicting
y_pred = clf.predict(X_test)
print("prediction done!!")
```

```python
#training
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(100, 50, 20,10), random_state=1)
clf.fit(X_train, y_train)
print("model trained--2")
#predicting
y_pred = clf.predict(X_test)
print("prediction done!!")
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(500, 175, 10), random_state=1)
clf.fit(X_train, y_train)
print("model trained--3")
#predicting
y_pred = clf.predict(X_test)
print("prediction done!!")
```


|model config|Accuracy on dev|

```python
# reference: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html
from sklearn.metrics import accuracy_score
print("accuracy on dev:")
print(accuracy_score(y_test, y_pred))
```
